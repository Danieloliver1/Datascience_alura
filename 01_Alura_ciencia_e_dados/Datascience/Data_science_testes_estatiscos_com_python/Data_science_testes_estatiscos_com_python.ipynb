```python
medias = list()
import numpy as np

np.random.seed(75243)

# sample aleatorio
temp = nota_media_dos_filmes_com_pelo_menos_10_votos.sample(frac=1)

# for i in range(1,len(temp))
#   medias.append(temp[0:i].mean())

medias = [temp[0:i].mean() for i in range(1,len(temp))]

plt.plot(medias)
```

Adiquirindo confiança

```python
from statsmodels.stats.weightstats import zconfint
zconfint(nota_media_dos_filmes_com_pelo_menos_10_votos)
```

```python
from scipy import stats

x = nota_media_dos_filmes_com_pelo_menos_10_votos
# Calculando a média e o desvio padrão
media = np.mean(x)
desvio_padrao = np.std(x)

# Tamanho da amostra
n = len(x)

# Calculando o intervalo de confiança da média com nível de confiança padrão de 95%
intervalo_confianca = stats.t.interval(0.95, df=n-1, loc=media, scale=desvio_padrao/np.sqrt(n))

print("Intervalo de Confiança da Média:", intervalo_confianca)
```

```python
from statsmodels.stats.weightstats import DescrStatsW

descr_todos_com_10_votos = DescrStatsW(nota_media_dos_filmes_com_pelo_menos_10_votos)

descr_todos_com_10_votos.tconfint_mean()
```

```python
# notas 1 média em relação ao tory story
np.random.seed(75241)

def calcula_teste(i):
  media = temp[0:i].mean()
  stat, p = ztest(temp[0:i], value  = 3.4320503405352594)
  return (i,media, p)


medias = [calcula_teste(i) for i in range(2, len(temp))]
valores = np.array([calcula_teste(i) for i in range(2, len(temp))])
medias
```
