```python
link1 = 'https://caelum-online-public.s3.amazonaws.com/2928-transformacao-manipulacao-dados/moveis_disponiveis.json'
link2 = 'https://caelum-online-public.s3.amazonaws.com/2928-transformacao-manipulacao-dados/dados_hospedagem.json'
```

```python
import pandas as pd

moveis_disponiveis = pd.read_json(link1)
dados_hospedagem = pd.read_json(link2)
```

```python
moveis_disponiveis.head()
```

```python
dados_hospedagem.head()
```

```python
# normalizando os dados json para visualizar como dataframe
dados = pd.json_normalize(dados_hospedagem['info_moveis'])
dados.head(5)
```

```python
colunas = list(dados.columns)
colunas
```

```python
dados = dados.explode(colunas[3:])
```

```python
dados.head()
```

```python
#resetando index
dados.reset_index(inplace=True,drop=True)
```

```python
dados.head()
```

```python
dados.info()
```

```python
import numpy as np
```

```python
# convertendo colunas
dados['max_hospedes'] = dados['max_hospedes'].astype(np.int64)
```

```python
dados.info()
```

```python
col_numericas = ['quantidade_banheiros','quantidade_quartos','quantidade_camas']
```

```python
dados[col_numericas] = dados[col_numericas].astype(np.int64)
```

```python
dados.info()
```

```python
dados['avaliacao_geral'] = dados['avaliacao_geral'].astype(np.float64)
```

```python
dados.info()
```

```python
dados['preco'] = dados['preco'].astype(np.float64)
```

```python
# .apply da para usar com um elemento ou seja coluna ou linha
dados['preco'] = dados['preco'].apply(lambda x: x.replace('$','').replace(',','').strip())
```

```python
dados['preco'].astype(np.float64)
```

```python
dados.info()
```

```python
# applymap() da para fazer com mais de uma coluna
dados[['taxa_deposito','taxa_limpeza']] = dados[['taxa_deposito','taxa_limpeza']].applymap(lambda x: x.replace('$','').replace(',','').replace('.','').strip())
```

```python
dados[['taxa_deposito','taxa_limpeza']] = dados[['taxa_deposito','taxa_limpeza']].astype(np.float64)
```

```python
# manipulando dados com str tipos string
dados['descricao_local'] = dados['descricao_local'].str.lower()
```

```python
dados['descricao_local'][3169]
```

```python
dados['descricao_local'] = dados['descricao_local'].str.replace('[^a-zA-Z0-9\-\']',' ',regex=True)
```

```python
dados['descricao_local'] = dados['descricao_local'].str.replace('(?<!\w)-(?!\w)',' ',regex=True)
```

```python
dados['descricao_local']
```

```python
# passando para uma lista
dados['descricao_local'] = dados['descricao_local'].str.split()
```

```python
dados['descricao_local']
```

```python
dados['comodidades'] = dados['comodidades'].str.replace('\{|}|\"',"",regex=True)
```

```python
dados['comodidades']
```

```python
# separada por virgula
dados['comodidades'] = dados['comodidades'].str.split(',')
```

```python
dados['comodidades']
```
