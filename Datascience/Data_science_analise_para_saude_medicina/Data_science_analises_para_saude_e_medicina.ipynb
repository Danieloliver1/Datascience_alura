```python
deaths = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/archived_data/archived_time_series/time_series_19-covid-Deaths_archived_0325.csv")
recovered = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/archived_data/archived_time_series/time_series_19-covid-Recovered_archived_0325.csv")
confirmed = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/archived_data/archived_time_series/time_series_19-covid-Confirmed_archived_0325.csv")

display(deaths.head())
display(recovered.head())
display(confirmed.head())
```

```python
confirmed.groupby('Country/Region').sum()[last_date].head()
```

```python
deaths.groupby('Country/Region').sum()[last_date].head()
```

```python
recovered.groupby('Country/Region').sum()[last_date].head()
```

```python
def latest_by_country(data):
  return data.groupby('Country/Region').sum().iloc[:,-1]

latest_by_country(confirmed)
```

```python
latest_by_country(recovered)
```

```python
latest_by_country(deaths)
```

```python
informations = [latest_by_country(confirmed),latest_by_country(recovered),latest_by_country(deaths)]
combined = pd.concat(informations, axis =1)
combined.columns = ['confirmed','deaths','recovered']
```

```python
combined
```

```python
# taxa de  letalidade Ã© expressa em %
# o nÃºmero de casos letais dentro dos casos que foram letais e nÃ£o letais

# imagina que existe uma doenÃ§a que em janeiro de 2020 afetou 100 pessoas
# 5 pessoas morreram (por exemplo). letalidade foi de 5%.
# nÃ£o digo que ela Ã© de 5% para sempre pq a doenÃ§a continuou existindo.
# e a doenÃ§a deixou de existir pela historia da humanidade.
```

```python
sum_up = combined.sum()
letality_rate_1 = sum_up['deaths'] / sum_up['confirmed']

display(letality_rate_1 * 100)
```

```python
sum_up
```

```python
letality_rate_2 = sum_up['deaths'] / (sum_up['recovered'] + sum_up['deaths'])
letality_rate_2
```

```python
print(letality_rate_1 * 100)
print(letality_rate_2 * 100)
```

```python
combined['letality_rate_1'] = letality_rate_1
combined['letality_rate_2'] = letality_rate_2
```

```python
def latest_by_country_at(data,date):
  return data.groupby('Country/Region').sum()[date]


informations = [latest_by_country_at(confirmed,'2/20/20'),latest_by_country_at(recovered,'2/20/20'),latest_by_country_at(deaths,'2/20/20')]
combined_2_20_20 = pd.concat(informations, axis =1)
combined_2_20_20.columns = ['confirmed','deaths','recovered']

sum_up = combined_2_20_20.sum()
letality_rate_1 = sum_up['deaths'] / sum_up['confirmed']
letality_rate_2 = sum_up['deaths'] / (sum_up['deaths'] + sum_up['recovered'])

print(letality_rate_1 * 100)
print(letality_rate_2 * 100)
```
