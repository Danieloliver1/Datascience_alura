```python
from textblob import TextBlob
```

```python
frase = 'Python é ótimo para achine Learning'
tb = TextBlob(frase)
```

```python
# verifica o setimento da frase vai de -1 a 1 quanto mais negativo significa negativo a frase e positivo o quanto é boa positivamente a frase.
tb.sentiment.polarity
```

```python
# traduzindo para o ingles
tb_en = tb.translate(from_lang='pt_br', to ='en')
```
